```r
Quadratic_Model_Age<-lm(Price~Age+I(Age^2), data=Honda_Accord)
summary(Quadratic_Model_Age)

plot(Price~Age, data=Honda_Accord)
lines(sort(Honda_Accord$Age), fitted(Quadratic_Model_Age)[order(Honda_Accord$Age)],type='b')
points(Honda_Accord$Price,Honda_Accord$Age,col='red')
```

## Would the fit improve significantly if you also included a cubic term? Justify your answer. 

Cubic Regression Model
